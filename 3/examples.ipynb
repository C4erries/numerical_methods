# Пример 1

**Задача Коши**

$$
\frac{dy}{dt}=y, \quad y(0)=1
$$
**Решение**
$$
y(t)=e^t
$$

```python
PROBLEM_NAME = "y' = y"

def f(t: float, y: np.ndarray) -> np.ndarray:
    # y' = y
    _ = t
    return np.asarray(y, dtype=float)

def exact(t: np.ndarray, t0: float, y0: np.ndarray) -> np.ndarray:
    # y(t) = y0 * exp(t - t0) (для скалярного y0)
    y0_arr = np.asarray(y0, dtype=float).reshape(-1)
    if y0_arr.size != 1:
        raise ValueError("Current exact() expects scalar y0.")
    t_arr = np.asarray(t, dtype=float)
    return (y0_arr[0] * np.exp(t_arr - t0)).reshape(-1, 1)

# in.txt
t0: 0
y0: 1
t_end: 2
h_values: 0.2 0.1
exact_h: 0.01
tol: 1e-12
max_iter: 40
```

**Задача Коши**

$$
\frac{dy}{dt}=y+t, \quad y(0)=1
$$
**Решение**
$$
y(t)=2e^t-t-1
$$

```python
PROBLEM_NAME = "y' = y + t"

def f(t: float, y: np.ndarray) -> np.ndarray:
    # y' = y + t
    return np.asarray(y, dtype=float) + t

def exact(t: np.ndarray, t0: float, y0: np.ndarray) -> np.ndarray:
    # y(t) = C*exp(t) - t - 1, C из y(t0)=y0
    y0_arr = np.asarray(y0, dtype=float).reshape(-1)
    if y0_arr.size != 1:
        raise ValueError("Current exact() expects scalar y0.")
    t_arr = np.asarray(t, dtype=float)
    c = y0_arr[0] + t0 + 1.0
    return (c * np.exp(t_arr - t0) - t_arr - 1.0).reshape(-1, 1)

# in.txt
t0: 0
y0: 1
t_end: 2
h_values: 0.2 0.1
exact_h: 0.01
tol: 1e-12
max_iter: 40
```

# Пример 3

**Задача Коши**

$$
\frac{dy}{dt}=y(1-y), \quad y(0)=0.1
$$
**Решение**
$$
y(t)=\frac{1}{1+9e^{-t}}
$$

```python
PROBLEM_NAME = "y' = y(1-y)"

def f(t: float, y: np.ndarray) -> np.ndarray:
    # y' = y(1-y)
    _ = t
    yv = np.asarray(y, dtype=float)
    return yv * (1.0 - yv)

def exact(t: np.ndarray, t0: float, y0: np.ndarray) -> np.ndarray:
    # y(t) = 1 / (1 + A*exp(-(t-t0))), A = (1-y0)/y0
    y0_arr = np.asarray(y0, dtype=float).reshape(-1)
    if y0_arr.size != 1:
        raise ValueError("Current exact() expects scalar y0.")
    if y0_arr[0] <= 0.0 or y0_arr[0] >= 1.0:
        raise ValueError("For this closed form, use 0 < y0 < 1.")
    t_arr = np.asarray(t, dtype=float)
    a = (1.0 - y0_arr[0]) / y0_arr[0]
    return (1.0 / (1.0 + a * np.exp(-(t_arr - t0)))).reshape(-1, 1)

# in.txt
t0: 0
y0: 0.1
t_end: 8
h_values: 0.2 0.1
exact_h: 0.01
tol: 1e-12
max_iter: 40
```

# Пример 4

**Задача Коши**

$$
\begin{cases}
x' = v,\\
v' = -x,\\
x(0)=1,\ v(0)=0
\end{cases}
$$
**Решение**
$$
x(t)=\cos t, \quad v(t)=-\sin t
$$

```python
PROBLEM_NAME = "x' = v, v' = -x"

def f(t: float, y: np.ndarray) -> np.ndarray:
    # y = [x, v]
    _ = t
    yv = np.asarray(y, dtype=float)
    return np.array([yv[1], -yv[0]], dtype=float)

def exact(t: np.ndarray, t0: float, y0: np.ndarray) -> np.ndarray:
    # x(t), v(t) для гармонического осциллятора
    y0_arr = np.asarray(y0, dtype=float).reshape(-1)
    if y0_arr.size != 2:
        raise ValueError("Current exact() expects y0=[x0, v0].")
    x0, v0 = y0_arr
    tau = np.asarray(t, dtype=float) - t0
    x = x0 * np.cos(tau) + v0 * np.sin(tau)
    v = -x0 * np.sin(tau) + v0 * np.cos(tau)
    return np.column_stack((x, v))

# in.txt
t0: 0
y0: 1 0
t_end: 10
h_values: 0.2 0.1
exact_h: 0.01
tol: 1e-12
max_iter: 40
```

# Пример 5

**Задача Коши**

$$
\begin{cases}
x' = v,\\
v' = -2v - 5x,\\
x(0)=1,\ v(0)=-1
\end{cases}
$$
**Решение**
$$
x(t)=e^{-t}\cos(2t), \quad v(t)=-e^{-t}\bigl(\cos(2t)+2\sin(2t)\bigr)
$$

```python
PROBLEM_NAME = "x' = v, v' = -2v - 5x"

def f(t: float, y: np.ndarray) -> np.ndarray:
    _ = t
    yv = np.asarray(y, dtype=float)
    return np.array([yv[1], -2.0 * yv[1] - 5.0 * yv[0]], dtype=float)

def exact(t: np.ndarray, t0: float, y0: np.ndarray) -> np.ndarray:
    y0_arr = np.asarray(y0, dtype=float).reshape(-1)
    if y0_arr.size != 2:
        raise ValueError("Current exact() expects y0=[x0, v0].")
    x0, v0 = y0_arr
    tau = np.asarray(t, dtype=float) - t0
    exp_tau = np.exp(-tau)
    cos_2tau = np.cos(2.0 * tau)
    sin_2tau = np.sin(2.0 * tau)
    a = x0
    b = 0.5 * (x0 + v0)
    x = exp_tau * (a * cos_2tau + b * sin_2tau)
    v = exp_tau * ((-a + 2.0 * b) * cos_2tau + (-b - 2.0 * a) * sin_2tau)
    return np.column_stack((x, v))

# in.txt
problem: ex5
t0: 0
y0: 1 -1
t_end: 8
h_values: 0.2 0.1
exact_h: 0.01
tol: 1e-12
max_iter: 50
```

# Пример 6

**Задача Коши**

$$
\begin{cases}
y_1' = y_1,\\
y_2' = y_1 + y_2,\\
y_3' = y_2 + y_3,\\
y_1(0)=1,\ y_2(0)=-1,\ y_3(0)=2
\end{cases}
$$
**Решение**
$$
y_1(t)=e^t, \quad y_2(t)=e^t(t-1), \quad y_3(t)=e^t\left(\frac{t^2}{2}-t+2\right)
$$

```python
PROBLEM_NAME = "y1' = y1, y2' = y1 + y2, y3' = y2 + y3"

def f(t: float, y: np.ndarray) -> np.ndarray:
    # y = [y1, y2, y3]
    _ = t
    yv = np.asarray(y, dtype=float)
    return np.array([yv[0], yv[0] + yv[1], yv[1] + yv[2]], dtype=float)

def exact(t: np.ndarray, t0: float, y0: np.ndarray) -> np.ndarray:
    # точное решение для 3-мерной линейной системы
    y0_arr = np.asarray(y0, dtype=float).reshape(-1)
    if y0_arr.size != 3:
        raise ValueError("Current exact() expects y0=[y10, y20, y30].")
    y10, y20, y30 = y0_arr
    tau = np.asarray(t, dtype=float) - t0
    exp_tau = np.exp(tau)
    y1 = y10 * exp_tau
    y2 = exp_tau * (y20 + y10 * tau)
    y3 = exp_tau * (y30 + y20 * tau + 0.5 * y10 * tau**2)
    return np.column_stack((y1, y2, y3))

# in.txt
problem: ex6
t0: 0
y0: 1 -1 2
t_end: 4
h_values: 0.2 0.1
exact_h: 0.01
tol: 1e-12
max_iter: 50
```
