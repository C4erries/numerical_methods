# Пример 2

**Краевая задача типа Эйлера-Коши**

$$
x^2y''+xy'-4y=x^3, \qquad x\in[1,2].
$$

В стандартном виде:

$$
y''+\frac{1}{x}y'-\frac{4}{x^2}y=x.
$$

Граничные условия:

$$
y(1)=0, \qquad y'(2)=0.
$$


Общее решение:

$$
y(x)=C_1x^2+\frac{C_2}{x^2}+\frac{x^3}{5}.
$$

Из граничных условий получаем

$$
y(x)=-\frac{49}{85}x^2+\frac{32}{85x^2}+\frac{x^3}{5}.
$$

```python
import numpy as np

PROBLEM_NAME = "medium"

def exact_scalar(x: np.ndarray) -> np.ndarray:
    x_arr = np.asarray(x, dtype=float)
    return -49.0 * x_arr * x_arr / 85.0 + 32.0 / (85.0 * x_arr * x_arr) + x_arr**3 / 5.0

def p(x: float) -> float:
    return 1.0 / x

def q(x: float) -> float:
    return -4.0 / (x * x)

def rhs(x: float) -> float:
    return x

def exact(x: np.ndarray, a: float, b: float, y_left: float, dy_right: float) -> np.ndarray:
    _ = a, b, y_left, dy_right
    return exact_scalar(x).reshape(-1, 1)


a: 1
b: 2
y_left: 0
dy_right: 0
n_values: 10 20 40
exact_n: 1000
boundary_order: 2

```

# Пример 3



$$
y''+\left(\sin(2x)+\frac{x}{2}\right)y'+(1+x^2)y=f(x), \qquad x\in[0,1].
$$

правая часть специально подобрана по известному точному решению:

$$
y(x)=e^{-3x}\sin(10x)+0.2\cos(5x)+x^2.
$$

Тогда правая часть равна

$$
\begin{aligned}
f(x)=&\ e^{-3x}(-60\cos(10x)-91\sin(10x))-5\cos(5x)+2 \\
&+\left(\sin(2x)+\frac{x}{2}\right)
\left(e^{-3x}(10\cos(10x)-3\sin(10x))-\sin(5x)+2x\right) \\
&+(1+x^2)\left(e^{-3x}\sin(10x)+0.2\cos(5x)+x^2\right).
\end{aligned}
$$



```python
PROBLEM_NAME = "hard"

def exact_scalar(x: np.ndarray) -> np.ndarray:
    x_arr = np.asarray(x, dtype=float)
    return np.exp(-3.0 * x_arr) * np.sin(10.0 * x_arr) + 0.2 * np.cos(5.0 * x_arr) + x_arr * x_arr

def exact_derivative(x: np.ndarray) -> np.ndarray:
    x_arr = np.asarray(x, dtype=float)
    return (
        np.exp(-3.0 * x_arr) * (10.0 * np.cos(10.0 * x_arr) - 3.0 * np.sin(10.0 * x_arr))
        - np.sin(5.0 * x_arr)
        + 2.0 * x_arr
    )

def exact_second_derivative(x: np.ndarray) -> np.ndarray:
    x_arr = np.asarray(x, dtype=float)
    return np.exp(-3.0 * x_arr) * (-60.0 * np.cos(10.0 * x_arr) - 91.0 * np.sin(10.0 * x_arr)) - 5.0 * np.cos(5.0 * x_arr) + 2.0

def p(x: float) -> float:
    return float(np.sin(2.0 * x) + 0.5 * x)

def q(x: float) -> float:
    return 1.0 + x * x

def rhs(x: float) -> float:
    return float(
        exact_second_derivative(x)
        + p(x) * exact_derivative(x)
        + q(x) * exact_scalar(x)
    )

def exact(x: np.ndarray, a: float, b: float, y_left: float, dy_right: float) -> np.ndarray:
    _ = a, b, y_left, dy_right
    return exact_scalar(x).reshape(-1, 1)


a: 0
b: 1
y_left: 0.2
dy_right: 2.6224308075507947
n_values: 80 160 320
exact_n: 2000
boundary_order: 2

```
