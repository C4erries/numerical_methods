# Конспект к задаче 4

## 1. Постановка задачи

Решается линейная краевая задача второго порядка

$$
y''(x)+p(x)y'(x)+q(x)y(x)=f(x), \qquad x\in[a,b].
$$

Граничные условия смешанные:

$$
y(a)=\alpha, \qquad y'(b)=\beta.
$$

Слева задано условие Дирихле, справа задано условие Неймана. После аппроксимации методом конечных разностей получается трехдиагональная система, которая решается методом прогонки.

## 2. Равномерная сетка

На отрезке $[a,b]$ вводится равномерная сетка

$$
x_i=a+ih, \qquad i=0,1,\dots,n,
$$

где

$$
h=\frac{b-a}{n}.
$$

Искомые значения обозначаются

$$
y_i\approx y(x_i).
$$

## 3. Разностная схема во внутренних узлах

Для внутренних узлов $i=1,\dots,n-1$ используются центральные разности:

$$
y''(x_i)\approx \frac{y_{i-1}-2y_i+y_{i+1}}{h^2},
$$

$$
y'(x_i)\approx \frac{y_{i+1}-y_{i-1}}{2h}.
$$

После подстановки в дифференциальное уравнение получаем

$$
A_i y_{i-1}+B_i y_i+C_i y_{i+1}=f_i,
$$

где

$$
A_i=\frac{1}{h^2}-\frac{p_i}{2h}, \qquad
B_i=-\frac{2}{h^2}+q_i, \qquad
C_i=\frac{1}{h^2}+\frac{p_i}{2h}.
$$

Здесь $p_i=p(x_i)$, $q_i=q(x_i)$, $f_i=f(x_i)$. 

## 4. Левое граничное условие Дирихле

Условие

$$
y(a)=\alpha
$$

записывается как первая строка системы:

$$
y_0=\alpha.
$$

То есть в матрице ставится единица на главной диагонали первой строки, а правая часть равна $\alpha$.

## 5. Правое граничное условие Неймана

Справа задано

$$
y'(b)=\beta.
$$

В программе доступны два варианта аппроксимации.

### 5.1. Первый порядок

$$
\frac{y_n-y_{n-1}}{h}=\beta.
$$

Отсюда последняя строка системы:

$$
-\frac{1}{h}y_{n-1}+\frac{1}{h}y_n=\beta.
$$

### 5.2. Второй порядок через фиктивный узел

По умолчанию используется более точный вариант. Вводится фиктивное значение $y_{n+1}$ и центральная формула

$$
\frac{y_{n+1}-y_{n-1}}{2h}=\beta.
$$

Тогда

$$
y_{n+1}=y_{n-1}+2h\beta.
$$

Подставляя это в исходное уравнение в точке $x_n=b$, получаем последнюю строку

$$
\frac{2}{h^2}y_{n-1}+\left(-\frac{2}{h^2}+q_n\right)y_n
=f_n-p_n\beta-\frac{2\beta}{h}.
$$

Этот вариант сохраняет второй порядок аппроксимации на правой границе.

## 6. Трехдиагональная система

После записи всех строк получается система

$$
M\mathbf y=\mathbf d,
$$

где матрица $M$ трехдиагональная. В программе она хранится не как полная матрица, а тремя массивами:

- `lower` - нижняя диагональ;
- `diagonal` - главная диагональ;
- `upper` - верхняя диагональ.

## 7. Диагональное преобладание

Для надежной работы метода прогонки желательно, чтобы трехдиагональная матрица имела диагональное преобладание:

$$
|b_i|\ge |a_i|+|c_i|.
$$

Если хотя бы в одной строке неравенство строгое и матрица не распадается на независимые блоки, то такая матрица обычно невырождена, а в прямом ходе прогонки не появляются нулевые знаменатели.

Для внутренних узлов нашей схемы

$$
a_i=\frac{1}{h^2}-\frac{p_i}{2h}, \qquad
b_i=-\frac{2}{h^2}+q_i, \qquad
c_i=\frac{1}{h^2}+\frac{p_i}{2h}.
$$

Если $|p_i|h\le 2$, то коэффициенты при соседних узлах не меняют знак и

$$
|a_i|+|c_i|=\frac{2}{h^2}.
$$

При этом диагональный коэффициент удовлетворяет

$$
|b_i|=\left|-\frac{2}{h^2}+q_i\right|.
$$

Отсюда видно важное условие: если $q_i\le 0$, то

$$
|b_i|=\frac{2}{h^2}-q_i\ge \frac{2}{h^2}=|a_i|+|c_i|.
$$

Значит, при достаточно малом шаге, $|p_i|h\le 2$, и при $q_i\le 0$ во внутренних строках есть диагональное преобладание. Если $q_i<0$, оно строгое.

Левая граничная строка $y_0=\alpha$ имеет диагональное преобладание автоматически, потому что в ней стоит только один коэффициент $1$.

Для правого условия Неймана ситуация зависит от выбранной аппроксимации. При первом порядке последняя строка

$$
-\frac{1}{h}y_{n-1}+\frac{1}{h}y_n=\beta
$$

имеет нестрогое диагональное преобладание: $|1/h|=|-1/h|$. При втором порядке через фиктивный узел последняя строка содержит коэффициенты

$$
a_n=\frac{2}{h^2}, \qquad b_n=-\frac{2}{h^2}+q_n.
$$

Поэтому при $q_n\le 0$ снова получается

$$
|b_n|\ge |a_n|.
$$

Итого: для типичного эллиптического случая с $q(x)\le 0$ и достаточно мелкой сеткой матрица разностной системы имеет диагональное преобладание. Если же $q(x)>0$ или член $p(x)y'$ велик на грубой сетке, диагональное преобладание может нарушиться; тогда прогонка всё еще может сработать, но это уже не гарантируется этим простым критерием.


## 8. Метод прогонки

Рассматривается система

$$
a_i y_{i-1}+b_i y_i+c_i y_{i+1}=d_i.
$$

Сначала выполняется прямой ход. Вводятся модифицированные коэффициенты

$$
c'_0=\frac{c_0}{b_0}, \qquad d'_0=\frac{d_0}{b_0}.
$$

Для следующих строк:

$$
m_i=b_i-a_i c'_{i-1},
$$

$$
c'_i=\frac{c_i}{m_i}, \qquad
d'_i=\frac{d_i-a_i d'_{i-1}}{m_i}.
$$

Затем выполняется обратный ход:

$$
y_n=d'_n,
$$

$$
y_i=d'_i-c'_i y_{i+1}, \qquad i=n-1,n-2,\dots,0.
$$

Метод работает за $O(n)$ операций, поэтому хорошо подходит именно для разностных схем с трехдиагональной матрицей.

## 9. Точность

Во внутренних узлах используются центральные разности второго порядка, поэтому локальная аппроксимация имеет порядок $O(h^2)$.

Если справа используется `boundary_order = 2`, то правое условие Неймана тоже согласовано со вторым порядком. В этом случае ожидается глобальная ошибка

$$
\max_i |y_i-y(x_i)| = O(h^2).
$$

Если выбрать `boundary_order = 1`, то ошибка на правой границе обычно снижает общий порядок до первого.

## 10. Как устроена программа

Задача разбита на модули:

- `solver.py` содержит построение разностной системы и метод прогонки;
- `input_parser.py` читает `in.txt`;
- `csv_io.py` пишет CSV-файлы;
- `plotter.py` строит графики;
- `main.py` задает конкретные функции `p`, `q`, `rhs`, `exact` и запускает весь сценарий.


## 11. Выходные файлы

После запуска `main.py` создаются:

- `exact_xy.csv` — точное решение на частой сетке;
- `num_n_*.csv` — численное решение для конкретного числа разбиений;
- `num_vs_exact_n_*.csv` — сравнение по узлам: `x`, `y_num`, `y_exact`, `delta`;
- `comparison.png` — общий график точного и численных решений;
- `errors.png` — график абсолютных погрешностей;
- `summary.csv` — таблица со средней и максимальной ошибкой, шагом и нормой невязки.
