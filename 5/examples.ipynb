# Пример 1

**Краевая задача**

$$
u_{xx}+u_{yy}=f(x,y), \qquad (x,y)\in[0,1]\times[0,1],
$$

$$
u(x,y)\big|_{\partial\Omega}=0.
$$

**Точное решение**

$$
u(x,y)=\sin(\pi x)\,\sin(\pi y).
$$

**Правая часть**

$$
f(x,y)=u_{xx}+u_{yy}=-2\pi^{2}\sin(\pi x)\sin(\pi y).
$$

````python
PROBLEM_NAME = "easy"

def exact_scalar(x: np.ndarray, y: np.ndarray) -> np.ndarray:
    x_arr = np.asarray(x, dtype=float)
    y_arr = np.asarray(y, dtype=float)
    return np.sin(np.pi * x_arr) * np.sin(np.pi * y_arr)

def rhs(x: float, y: float) -> float:
    return -2.0 * float(np.pi ** 2 * np.sin(np.pi * x) * np.sin(np.pi * y))

def exact_value(x, y) -> np.ndarray:
    return exact_scalar(x, y)

def boundary(x: float, y: float) -> float:
    return float(exact_value(x, y))

def exact_grid(x: np.ndarray, y: np.ndarray) -> np.ndarray:
    xx, yy = np.meshgrid(np.asarray(x, dtype=float), np.asarray(y, dtype=float))
    return exact_value(xx, yy)


x0: 0
x1: 1
y0: 0
y1: 1
n_values: 8 16 32
exact_n: 160
method: seidel
tol: 1e-10
max_iter: 10000
````

# Пример 2

**Краевая задача**

$$
u_{xx}+u_{yy}=f(x,y), \qquad (x,y)\in[0,1]\times[0,1],
$$

$$
u(x,y)\big|_{\partial\Omega}=\bigl(x^{2}-y^{2}\bigr)\,e^{x+y}.
$$

**Точное решение**

$$
u(x,y)=\bigl(x^{2}-y^{2}\bigr)\,e^{x+y}.
$$

**Правая часть**

Прямой дифференциацией:

$$
u_{xx}=\bigl(2+4x+x^{2}-y^{2}\bigr)e^{x+y}, \qquad
u_{yy}=\bigl(-2-4y+x^{2}-y^{2}\bigr)e^{x+y}.
$$

Тогда

$$
f(x,y)=u_{xx}+u_{yy}=4\,(x-y)\,e^{x+y}.
$$

Граничные значения берутся из точного решения на сторонах квадрата.

````python
import numpy as np

PROBLEM_NAME = "medium"

def exact_scalar(x: np.ndarray, y: np.ndarray) -> np.ndarray:
    x_arr = np.asarray(x, dtype=float)
    y_arr = np.asarray(y, dtype=float)
    return (x_arr * x_arr - y_arr * y_arr) * np.exp(x_arr + y_arr)

def rhs(x: float, y: float) -> float:
    return float(4.0 * (x - y) * np.exp(x + y))

def exact_value(x, y) -> np.ndarray:
    return exact_scalar(x, y)

def boundary(x: float, y: float) -> float:
    return float(exact_value(x, y))

def exact_grid(x: np.ndarray, y: np.ndarray) -> np.ndarray:
    xx, yy = np.meshgrid(np.asarray(x, dtype=float), np.asarray(y, dtype=float))
    return exact_value(xx, yy)


x0: 0
x1: 1
y0: 0
y1: 1
n_values: 10 20 40
exact_n: 200
method: seidel
tol: 1e-10
max_iter: 20000
````

# Пример 3

**Краевая задача**

$$
u_{xx}+u_{yy}=f(x,y), \qquad (x,y)\in[0,1]\times[0,1],
$$

граница берётся из точного решения

$$
u(x,y)\big|_{\partial\Omega}=u_{\text{exact}}(x,y).
$$

**Точное решение** (быстро осциллирующее, со смешением масштабов)

$$
u(x,y)=e^{-3x}\sin(10x)\cos(8y)+0.2\,\cos(5y)+x^{2}y.
$$

**Правая часть**

Вторые частные производные:

$$
u_{xx}(x,y)=e^{-3x}\bigl(-91\sin(10x)-60\cos(10x)\bigr)\cos(8y)+2y,
$$

$$
u_{yy}(x,y)=-64\,e^{-3x}\sin(10x)\cos(8y)-5\cos(5y).
$$

Тогда

$$
f(x,y)=u_{xx}+u_{yy}=e^{-3x}\cos(8y)\bigl(-155\sin(10x)-60\cos(10x)\bigr)+2y-5\cos(5y).
$$

В коде правая часть вычисляется не через свёрнутое выражение, а как сумма
двух частных производных — так удобнее отлаживать и менять формулы.

````python
PROBLEM_NAME = "hard"

def exact_scalar(x: np.ndarray, y: np.ndarray) -> np.ndarray:
    x_arr = np.asarray(x, dtype=float)
    y_arr = np.asarray(y, dtype=float)
    return (
        np.exp(-3.0 * x_arr) * np.sin(10.0 * x_arr) * np.cos(8.0 * y_arr)
        + 0.2 * np.cos(5.0 * y_arr)
        + x_arr * x_arr * y_arr
    )

def exact_u_xx(x: np.ndarray, y: np.ndarray) -> np.ndarray:
    x_arr = np.asarray(x, dtype=float)
    y_arr = np.asarray(y, dtype=float)
    return (
        np.exp(-3.0 * x_arr)
        * (-91.0 * np.sin(10.0 * x_arr) - 60.0 * np.cos(10.0 * x_arr))
        * np.cos(8.0 * y_arr)
        + 2.0 * y_arr
    )

def exact_u_yy(x: np.ndarray, y: np.ndarray) -> np.ndarray:
    x_arr = np.asarray(x, dtype=float)
    y_arr = np.asarray(y, dtype=float)
    return (
        -64.0 * np.exp(-3.0 * x_arr) * np.sin(10.0 * x_arr) * np.cos(8.0 * y_arr)
        - 5.0 * np.cos(5.0 * y_arr)
    )

def rhs(x: float, y: float) -> float:
    return float(exact_u_xx(x, y) + exact_u_yy(x, y))

def exact_value(x, y) -> np.ndarray:
    return exact_scalar(x, y)

def boundary(x: float, y: float) -> float:
    return float(exact_value(x, y))

def exact_grid(x: np.ndarray, y: np.ndarray) -> np.ndarray:
    xx, yy = np.meshgrid(np.asarray(x, dtype=float), np.asarray(y, dtype=float))
    return exact_value(xx, yy)


x0: 0
x1: 1
y0: 0
y1: 1
n_values: 40 80 160
exact_n: 800
method: seidel
tol: 1e-10
max_iter: 50000
````

# Интерактивная 3D-визуализация

Помимо тепловых карт `comparison.png` и `errors.png`, `main.py` строит
интерактивный HTML `out/surface.html` (через `plotly`).

В нём можно:
- вращать камеру мышью, приближать/отдалять колесом;
- слайдером внизу выбирать, что отображать:
  - `exact` — точное решение;
  - `num + exact, n=...` — численное решение поверх полупрозрачного точного (видно расхождение);
  - `num only, n=...` — только численное;
  - `|error|, n=...` — модуль ошибки в каждой точке сетки.

Запуск:
```bash
pip install plotly
python3 main.py --no-show
xdg-open out/surface.html
```
