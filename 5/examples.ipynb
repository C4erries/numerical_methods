## Пример 1 (полином)

**f(x):** $x^2+1$

**Уравнение:** $\int_0^x (t^2+1)dt = 4/3$

Ожидаемо: `x = 1`.

```python
# --- in.txt ---
'''
0 1.333333333333333 2 2000 1e-10 50
'''

def f(x: float) -> float:
    return x**2 + 1.0
```

## Пример 2 (тригонометрия + сдвиг вверх, f(x) > 0)

**f(x):** $\sin(x)+2$

**Уравнение:** $\int_0^x (\sin t + 2)dt = 3.929262798332297$

Ожидаемо: `x = 1.5`.

```python
# --- in.txt ---
'''
0 3.929262798332297 3 4000 1e-10 50
'''

def f(x: float) -> float:
    return math.sin(x) + 2.0
```

## Пример 3 (экспонента)

**f(x):** $e^x$

**Уравнение:** $\int_0^x e^t dt = e-1$

Ожидаемо: `x = 1`.

```python
# --- in.txt ---
'''
0 1.718281828459045 2 5000 1e-10 50
'''

def f(x: float) -> float:
    return math.exp(x)
```

## Пример 4 (рациональная функция)

**f(x):** $\frac{1}{1+x^2}$

**Уравнение:** $\int_0^x \frac{1}{1+t^2}dt = \pi/4$

Ожидаемо: `x = 1`.

```python
# --- in.txt ---
'''
0 0.785398163397448 2 5000 1e-10 50
'''

def f(x: float) -> float:
    return 1.0 / (1.0 + x**2)
```

## Пример 5 (корень, ограничение по области определения)

**f(x):** $\sqrt{x+2}$ (требуется $x \ge -2$)

**Уравнение:** $\int_{-1}^x \sqrt{t+2}dt = 2.797434948471088$

Ожидаемо: `x = 1`.

```python
# --- in.txt ---
'''
-1 2.797434948471088 2 5000 1e-10 50
'''

def f(x: float) -> float:
    return math.sqrt(x + 2.0)
```
