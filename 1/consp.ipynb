```markdown
## Метод вращений Якоби (для симметричной полной задачи собственных значений)


### 1) Постановка задачи
Дана **вещественная симметричная** матрица
$$A=A^T\in\mathbb{R}^{n\times n}.$$
Найти собственные значения и собственные векторы:
$$Ax=\lambda x.$$

Для симметричной матрицы существует ортогональная матрица собственных векторов $X$ такая, что
$$X^TAX=\Lambda,\qquad \Lambda=\operatorname{diag}(\lambda_1,\dots,\lambda_n).$$

Это и есть цель метода: построить последовательность ортогональных преобразований, которые зануляют внедиагональные элементы

---


### 2) Преобразование подобия
Матрицы $A$ и $B$ называются **подобными**, если
$$B=C^{-1}AC,$$
где $C$ — невырожденная матрица.

**Свойство:** преобразование подобия сохраняет спектр (собственные значения).  
Если $\{\lambda,x\}$ — собственная пара $B$, то $\{\lambda, Cx\}$ — собственная пара $A$.

В методе Якоби берутся **ортогональные** преобразования ($C^T=C^{-1}$), т.е.
$$B = T^TAT,$$
поэтому:
- собственные значения сохраняются,
- симметрия сохраняется: если $A=A^T$, то $B=B^T$.

---


### 3) Матрица плоского вращения $T_{ij}$ 
Берём пару индексов $(i,j)$, $i\neq j$, и строим матрицу поворота Гивенса (плоского вращения) $T_{ij}$ 
- на подпространстве $(i,j)$ она действует как 2D-поворот:
$$
\tilde T=
\begin{pmatrix}
c & s\\
-s & c
\end{pmatrix},
\qquad c^2+s^2=1.
$$
Обычно $c=\cos\alpha,\ s=\sin\alpha$.

Матрица $T_{ij}$ **ортогональна**

$B$ подобна $A$.:
$$B=T_{ij}^TAT_{ij} \space(4.29)$$


---


### 4) Якоби итерации
Строится последовательность:
$$B_0=A,$$
$$B_{k+1}=T_{i_k j_k}^T\,B_k\,T_{i_k j_k}.$$

На каждом шаге выбирают пару $(i,j)$ и угол (то есть $c,s$) так, чтобы занулить элемент $b_{ij}$ (а значит и $b_{ji}$, всилу симметрии):
$$ (B_{k+1})_{ij}=0.$$

В пределе $B_k\to \Lambda$ (диагональная), а накопленное произведение поворотов даёт собственные векторы.

---


### 5) Один шаг 
#### 5.1) Условие зануления $a_{ij}$ и формула для угла
Рассматриваем 2×2-блок по индексам $(i,j)$:
$$
\tilde A=
\begin{pmatrix}
a_{ii} & a_{ij}\\
a_{ij} & a_{jj}
\end{pmatrix}.
$$

**Одним ортогональным поворотом невозможно одновременно занулить оба внедиагональных элемента, если $a_{ij}\neq a_{ji}$.**

Выбираем $\alpha$ так, чтобы после преобразования
$\tilde B=\tilde T^T\tilde A\tilde T$
получилось $\tilde b_{12}=0$.

Получается:
$$\tan(2\alpha)=\frac{2a_{ij}}{a_{ii}-a_{jj}},\qquad \alpha\in\left(-\frac{\pi}{4},\frac{\pi}{4}\right].$$

#### 5.2) Расчёт $c,s$ 
В реализации используется устойчивая схема (en.wiki):
$$\tau = \frac{a_{jj}-a_{ii}}{2a_{ij}},$$
$$t = \frac{\operatorname{sign}(\tau)}{|\tau|+\sqrt{1+\tau^2}},$$
$$c = \frac{1}{\sqrt{1+t^2}},\qquad s = t c.$$

Если $a_{ij}=0$, вращение на этом шаге не выполняется.

#### 5.3) Формулы обновления матрицы
После выбора $c,s$ строится $B=T_{ij}^TAT_{ij}$.

**Новые диагональные элементы:**
$$b_{ii}=c^2a_{ii}+s^2a_{jj}+2csa_{ij},$$
$$b_{jj}=s^2a_{ii}+c^2a_{jj}-2csa_{ij}.$$

**Зануляем:**
$$b_{ij}=b_{ji}=0.$$

**Для всех $m\neq i,j$ (формула (4.31)):**
$$b_{im}=b_{mi}=c\,a_{im}+s\,a_{jm},$$
$$b_{jm}=b_{mj}=-s\,a_{im}+c\,a_{jm}.$$

**Остальные элементы не меняются.**

---


### 6) Почему метод сходится
Обозначим внедиагональную часть как матрицу $V_A$ (все диагональные элементы заменены на нули).  
Тогда квадрат её нормы Фробениуса:
$$\|V_A\|_F^2=\sum_{i\neq j}a_{ij}^2.$$

Если на шаге мы выбрали $(i,j)$ и занулили $a_{ij}$, то:
$$\|V_B\|_F^2=\|V_A\|_F^2-2a_{ij}^2. (4.32)$$
$$|a_{ij}| = \max{|a_{ml}|} \Longrightarrow a_{ij}^2 \geq \frac{1}{n(n-1)}\|V_A\|_F^2$$
$$\|V_{B_k}\|_F^2 \leq (1-\frac{2}{n(n-1)})\|V_{B_{k-1}}\|_F^2 \leq\dots\leq (1-\frac{2}{n(n-1)})^k\|V_A\|_F^2 \xrightarrow[k\rightarrow\infty]{} 0$$
Это и даёт сходимость к диагональной матрице.

---


### 7) Выбор ключевого элемента $(i,j)$
**Классический Якоби:** на каждом шаге брать максимальный по модулю внедиагональный элемент:
$$|a_{ij}|=\max_{m<\ell}|a_{m\ell}|.$$
Плюс: гарантия уменьшения $\|V\|_F$.  
Минус: поиск максимума стоит $O(n^2)$ на шаг.

---


### 8) собственные векторы (накопление вращений)
Пусть на шагах применялись вращения $T_{i_0j_0},\dots,T_{i_{k-1}j_{k-1}}$.  
Тогда
$$B_k = T_k^TAT_k,$$
где
$$T_k = T_{i_0j_0}\,T_{i_1j_1}\cdots T_{i_{k-1}j_{k-1}}.$$

- старт: $X=I$,
- на каждом шаге поворачиваем столбцы $i$ и $j$ так же, как в матрице $A$:  $X = X\,T_{ij}$.

В конце:
- собственные значения: $\lambda_i \approx (B_k)_{ii}$,
- собственные векторы: $x_i$ — $i$-й столбец $X$.

---
